```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```python
df_1 = pd.read_excel('LinkedIn-nigerian-data-section-1.xlsx')
df_2 = pd.read_excel('LinkedIn-nigerian-data-section-4.xlsx')
df = pd.concat([df_1, df_2])
```

```python
words = ''
for post in df['textContent']:
    post = str(post).lower()
    
    tokens = post.split()
    words += " ".join(tokens)+" "
```

```python
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords = stopwords).generate(words)
  
                      
plt.figure(figsize = (8, 8))
plt.imshow(wordcloud)
```
